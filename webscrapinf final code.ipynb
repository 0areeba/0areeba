```python
import requests
import json
import re
import regex
from bs4 import BeautifulSoup


res = requests.get('https://foursquare.com/foursquare?all=lists&geoId=72057594043056517')

soup = BeautifulSoup(res.text, 'html.parser')

list_scripts  = soup.find_all('script', type='text/javascript')
data = ''
for script in list_scripts:
  if 'userprofile2' in script.text: 
    data = script.text

pattern = regex.compile(r'\{(?:[^{}]|(?R))*\}')
json_data_string = pattern.findall(data.split('user:',1)[1])
data = json.loads(json_data_string[0])

with open('E:\\jupyter\\foursquare.json', 'w') as f:
  # f.write(json_data_string[0])
  json.dump(data, f)
    

df = pd.read_json (r'E:\jupyter\foursquare.json')
df.to_csv (r'E:\jupyter\web_scrap_csv.csv', index = None)
    
    
    
```
